```python
nums = []
for i in range(len(ba) // 16):
    # REAL
    start = 16 * i
    end = start + 8
    
    bits = ba[start: end]
    real = struct.unpack('d', bits)[0]
    
    start = 16 * i + 8
    end = 16 * i + 16
    imag = struct.unpack('d', bits)[0]
    nums.append(complex(real, imag))
```

```python
import numpy as np
import matplotlib.pyplot as plt

nums = np.array(nums)


plt.scatter(nums.real, nums.imag)
plt.show()
```

```python
import numpy as np

Bytes = np.fromfile("input_clean", dtype = "uint8")
Bits_original = np.unpackbits(Bytes)

Bytes = np.fromfile('output', dtype='uint8')
Bits_decoded = np.unpackbits(Bytes)

```

```python
samples = 10**4
start = 0

def correlate(a, b,length, lag=0):
    return sum([a[i]*b[(i + lag) % length] for i in range(length)])


result = correlate(Bits_original, Bits_original, samples)
autoCorrInput = np.sqrt(result)

crossCor = max([correlate(Bits_original, Bits_decoded, samples, lag=start) for start in range(samples//2)])
lag = np.argmax([correlate(Bits_original, Bits_decoded, samples, lag=start) for start in range(samples//2)])

result = correlate(Bits_decoded[lag:], Bits_decoded[lag:], samples)
autoCorrOutput = np.sqrt(result)

print(lag)
print(autoCorrInput, autoCorrOutput, crossCor)
print("Correlation:" , crossCor / (autoCorrInput * autoCorrOutput))
```

```python
# Now lets work with some noise
import struct
with open('input_clean', 'rb') as inFile:
    ba = bytearray(inFile.read())
    nums_original_clean = [int(bit) for bit in ba]

with open('output_noisy', 'rb') as inFile:
    ba = bytearray(inFile.read())
    
nums_noisy = []
nums_decoded_noisy = []

for i in range(len(ba) // 8):
    num = struct.unpack('d', ba[8*i: 8*i + 8])[0]
    if num > 0:
        nums_decoded_noisy.append(1)
    else:
        nums_decoded_noisy.append(0)
        
samples = 1000

result = np.correlate(nums_original_clean[:samples], nums_original_clean[:samples], mode='full')
autoCorr = result[result.size//2:][0]
crossCor = max(np.correlate(nums_original_clean[:samples], nums_decoded_noisy[:samples], mode='full'))
print(autoCorr, crossCor)
print("Correlation:" , crossCor * 100 / autoCorr)
```
